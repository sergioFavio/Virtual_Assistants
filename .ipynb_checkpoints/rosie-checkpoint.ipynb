```python
from gtts import gTTS
import speech_recognition as sr
#from subprocess import call     # MAC / LINUX
from playsound import playsound # WINDOWS
from requests import get
from bs4 import BeautifulSoup
import webbrowser as browser
import json
import os

##### CONFIGURACIONES #####
hotword = 'camila'
#hotword = 'athenea'

#with open('camilapythonasistente-7723e4c012dd.json') as credenciais_google:
with open('asistentepython2-2b41fbcac156.json') as credenciais_google:
    credenciais_google = credenciais_google.read()


##### FUNCIONES PRINCIPALES #####

def monitora_audio():
    microfone = sr.Recognizer()
    with sr.Microphone() as source:
        while True:
            print("Aguardando o Comando: ")
            audio = microfone.listen(source)
            try:
                trigger = microfone.recognize_google_cloud(audio, credentials_json=credenciais_google, language='pt-br')
                trigger = trigger.lower()

                if hotword in trigger:
                    print('COMANDO: ', trigger)
                    responde('feedback')
                    executa_comandos(trigger)
                    break

            except sr.UnknownValueError:
                print("Google no entiende ese audio")
            except sr.RequestError as e:
                print("Could not request results from Google Cloud Speech service; {0}".format(e))
    return trigger

def responde(arquivo):
    #call(['afplay', 'audios/' + arquivo + '.mp3']) # OSC
    playsound('audios/' + arquivo + '.mp3')        # WINDOWS

def cria_audio(mensagem):
    os.remove('audios/mensagem.mp3')
    #tts = gTTS(mensagem, lang='es-us')
    #tts = gTTS(mensagem, lang='de')
    #tts = gTTS(mensagem, lang='ru') # voz masculina
    #tts = gTTS(mensagem, lang='cs')  # checo
    #tts = gTTS(mensagem, lang='ko') # voz masculina
    #tts = gTTS(mensagem, lang='ar')
    #tts = gTTS(mensagem, lang='zh-tw')
    #tts = gTTS(mensagem, lang='ja')
    #tts = gTTS(mensagem, lang='de') # humorista femenina
    #tts = gTTS(mensagem, lang='it')
    #tts = gTTS(mensagem, lang='pt-br')
    tts = gTTS(mensagem, lang='fr') #asesora en moda y elegancia Facha Michelle Pituque ...
    tts.save('audios/mensagem.mp3')
    print('CAMILA: ', mensagem)
    playsound('audios/mensagem.mp3')        # WINDOWS
    #call(['afplay', 'audios/mensagem.mp3']) # OSX
    


def executa_comandos(trigger):
    if 'notícias' in trigger:
        ultimas_noticias()

    elif 'tocar' in trigger and 'bee gees' in trigger:
        playlists('bee_gees')

    elif 'tocar' in trigger and 'taylor davis' in trigger:
        playlists('taylor_davis')
        
    elif 'tocar' in trigger and 'trio los angeles' in trigger:
        playlists('trio_los_angeles')
        
    elif 'tocar' in trigger and 'mambolê' in trigger:
        playlists('mambole')
        
    elif 'tocar' in trigger and 'josé augusto' in trigger:
        playlists('jose_augusto')     

    elif 'tempo agora' in trigger:
        previsao_tempo(tempo=True)

    elif 'temperatura hoje' in trigger:
        previsao_tempo(minmax=True)

    else:
        mensagem = trigger.strip(hotword)
        cria_audio(mensagem)
        print('COMANDO INVALIDO', mensagem)
        responde('comando_invalido')


##### FUNCIONES COMANDOS #####

def ultimas_noticias():
    site = get('https://news.google.com/news/rss?ned=pt_br&gl=BR&hl=pt')
    noticias = BeautifulSoup(site.text, 'html.parser')
    for item in noticias.findAll('item')[:5]:
        #mensagem = item.title.text
        #mensagem='Ciao, soy Giuseppe Pomarola, cocinero italiano, e vamos a enseñar a cocinar comida italiana!, pronto en radio estación de arica, chile. arrivederci'
        #cria_audio(mensagem)
        #mensagem='Lasagna, Aunque la preparación de la lasaña varía según la región en la que se prepare consiste en un tipo de pasta con forma de lámina cuadrada que se coloca en forma de bloque y se rellena con carne, pescado o verdura. Entre lámina y lámina también lleva bechamel, y para terminar abundante queso con el que se le dará su aspecto habitual gratinado' 
        #cria_audio(mensagem)
        #mensagem='hola, Bonjour, soy Michelle Pituqué y vamos hablar de moda, elegancia y charme' 
        #cria_audio(mensagem)
        #mensagem='Soy Anke Lachen, humorista alemana, i este es el chiste del día en radio estación de arica - chile: No entiendo porque no consigo pareja. ¿ya probaste quitándote el bigote? -Aún no, pues hazlo ahora, Paola' 
        #mensagem='lo siento, mi trabajo es solo contar chistes'
        #mensagem='scusa, mi trabajo es solo cucinare comida italiana'
        #mensagem='une momento por favor, merci'
        mensagem='lo siento, pero eso no, es mi trabajo'
        cria_audio(mensagem)


def playlists(album):
    if album == 'bee_gees':
        browser.open('https://open.spotify.com/track/33ALuUDfftTs2NEszyvJRm')
    elif album == 'taylor_davis':
        browser.open('https://open.spotify.com/track/3MKep4BfEwSlAHuFJrA9aV')
    elif album == 'trio_los_angeles':
        browser.open('https://www.youtube.com/watch?v=LMWMX43AvEA&list=PLYYZsFVm6afjy2NpPA2UYsiOjUOkNEwtm')
    elif album == 'mambole':
        browser.open('https://www.youtube.com/watch?v=YKfp_-ywEpA')   
    elif album == 'jose_augusto':
        browser.open('https://www.youtube.com/watch?v=jvJy7VieqeI')

def previsao_tempo(tempo=False, minmax=False):
    site = get('http://api.openweathermap.org/data/2.5/weather?q=rio de janeiro&appid=a1f0d2c3fb0cbfb7d3230d60967b95a2&units=metric&lang=pt')
    clima = site.json()
    #print(json.dumps(clima, indent=4))
    temperatura=clima['main']['temp']
    minima=clima['main']['temp_min']
    maxima=clima['main']['temp_max']
    descricao=clima['weather'][0]['description']
    cidade=clima['name']
    
    if tempo:
        mensagem = f'No momento fazem {temperatura} graus com: {descricao} em {cidade}'
    if minmax:
        mensagem = f'Minima de {minima} e maxima de {maxima} graus em {cidade}'
    cria_audio(mensagem)

def main():
    while True:
        monitora_audio()

main()
```
